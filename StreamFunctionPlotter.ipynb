# Visualising Streamfunctions at Fixed Time

In fluid mechanics, a streamfunction, denoted as  $\Psi$, is a scalar function used to describe two-dimensional flow. The velocity components $u$ and $v$ in the $x$ and $y$ directions are given by the derivative of the streamfunction in the y and x directions respectively.

The streamfunction simplifies visualization and analysis of fluid flow by plotting its contours to form streamlines, which show the flow direction and structure. Streamlines help identify features like vortices and recirculation zones, and the streamfunction inherently satisfies the continuity equation for incompressible flows, making it a valuable tool in fluid mechanics.

The first cell plots a contour map of a streamfunction, with no option for seeing time evolution. The second cell plots a 5x1 subplot for various values of time that can be helpful if you want to see how flows evolve with time. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range and resolution of x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a grid of x and y values
X, Y = np.meshgrid(x, y)

# Define the time parameter t
t = 0  # You can change the value of t if needed
K = 0.8
k = 10
# Calculate the stream function psi
psi = np.cosh(K*X) * np.cos(k*Y)-np.sinh(K*Y) * np.sin(k*X)

# Plot the stream function
plt.figure(figsize=(8, 6))
plt.contour(X, Y, psi, levels=20, cmap='viridis')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stream Function: psi = cosh(Kx)sin(ky) - sinh(Kx)cos(ky)')
plt.colorbar(label='Psi')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range and resolution of x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Create a grid of x and y values
X, Y = np.meshgrid(x, y)

# Define the values of t
t_values = [0, 1, 2, 3, 4]

# Create subplots for each value of t
plt.figure(figsize=(7, 30))
for i, t in enumerate(t_values, 1):
    plt.subplot(5, 1, i)
    psi = np.cos(X - t) * np.sin(Y)
    plt.contour(X, Y, psi, levels=20, cmap='viridis')
    plt.xlabel('x', font = 'serif')
    plt.ylabel('y', font = 'serif')
    plt.title(f't = {t}', font = 'serif')
    plt.colorbar(label='Psi')
    plt.grid(True)

plt.suptitle('Streamfunction: psi = cos(x-t)sin(y)', fontsize=16, font = 'serif')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
#plt.savefig('streamfunctions', bbox_inches = 'tight', dpi = 400)
plt.show()
```
